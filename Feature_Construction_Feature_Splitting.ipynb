```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Titanic wala dataset.csv')[['Age','Pclass','SibSp','Parch','Survived']]
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=5))
```

# **Applying feature Construction**

```python
X['Familly_size'] = X['SibSp'] + X['Parch'] + 1
X.head()
```

```python
def myfunc(num):
      if num == 1:
        #Alone
        return 0
      elif num > 1 and num <= 4:
        #small familly
         return 1
      else:
        #large familly
        return 2
```

```python
myfunc(4)
```

```python
X['Familly_type'] = X['Familly_size'].apply(myfunc)
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Familly_size'],inplace=True)
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=5))
```
